## Описание задачи от заказчика

**Задача**: Автоматическая классификация обращений граждан

**Авторство задачи**: Рязанская область.

**Оригинальное название**: "Объективная оценка степени удовлетворенности населения"

**Краткое описание проблемы**: Существует множество разрозненных источников поступления обратной связи от населения жалобы, предложения . Заявки поступают в различном формате в различные инстанции, которые, в свою очередь, не всегда компетентны по поступающему вопросу. Поступающая информация часто теряется, что ведет к нерешению проблем, росту однотипных повторяющихся заявок от населения и, как следствие, к повышению неудовлетворенности населения.

**Постановка задачи**
Создать программу с машинным обучением, которая на основании содержания текста обращения могла бы определять тему обращения и ответственный орган власти.

**Данные**
Представленный датасет содержит уже предразмеченные человеком обращения.
Входные данные содержатся в поле text.

**Ожидаемый результат**
Разрабатываемый продукт должен генерировать:

- *theme* - генерируемый текст, содержащий ключевые факты из текст обращения

- *category* - предложение по отнесению обращения к определенной категории из справочника.

- *executor* - предполагаемый исполнитель, которому должно быть адресовано обращение.

В результате может быть создана система приема обращений граждан, которая автоматически направляет обращение непосредственному исполнителю.

### Описание решения

#### Шаги решения/подходы
1. Разведка данных (исследование входных данных)
 * что из себя представляют данные
 * структура данных, распределение по целевым признакам
 * количественные и качественные характеристики
 * какие присутствуют зависимости
2. Чистка данных и подготовка данных
 * удаление очевидных шумов (например, *тестовый исполнитель*, пустые обращения граждан и проч.)
 * удаление неподходящих данных (присутствие этих данных ухудшает качество предсказаний)
    - слишком длинные обращения
    - уменьшение размерности целевых признаков (уменьшение количества предсказываемых категорий, тем, исполнителей)
 * создание признаков: Bag of Word, TF-IDF слов/н-грам
 * формирование выборок для обучения классификаторов
3. Поиск классификаторов
 * обучение, сравнение качества, отбор классификаторов
 * повторная работа с данными (чистка и проч.)

````python
def plot_labels(rects, ax, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')


        
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
#     plt.colorbar()
    plt.grid(False)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Истинное')
    plt.xlabel('Прогноз')
    
    

def get_most_important_features(vectorizer, model, n=5):
    index_to_word = {v:k for k,v in vectorizer.vocabulary_.items()}
    
    # loop for each class
    classes ={}
    for class_index in range(model.coef_.shape[0]):
        word_importances = [(el, index_to_word[i]) for i,el in enumerate(model.coef_[class_index])]
        sorted_coeff = sorted(word_importances, key = lambda x : x[0], reverse=True)
        tops = sorted(sorted_coeff[:n], key = lambda x : x[0])
        bottom = sorted_coeff[-n:]
        classes[class_index] = {
            'tops':tops,
            'bottom':bottom
        }
    return classes



def plot_important_words(top_scores, top_words, bottom_scores, bottom_words, name):
    y_pos = np.arange(len(top_words))
    top_pairs = [(a,b) for a,b in zip(top_words, top_scores)]
    top_pairs = sorted(top_pairs, key=lambda x: x[1])
    
    bottom_pairs = [(a,b) for a,b in zip(bottom_words, bottom_scores)]
    bottom_pairs = sorted(bottom_pairs, key=lambda x: x[1], reverse=True)
    
    top_words = [a[0] for a in top_pairs]
    top_scores = [a[1] for a in top_pairs]
    
    bottom_words = [a[0] for a in bottom_pairs]
    bottom_scores = [a[1] for a in bottom_pairs]
    
    fig = plt.figure(figsize=(7, 4))  

    plt.subplot(121)
    plt.barh(y_pos,bottom_scores, align='center', alpha=0.5)
    plt.title('Нерелевантные', fontsize=12)
    plt.yticks(y_pos, bottom_words, fontsize=14)
    plt.suptitle('Key words', fontsize=16)
    plt.xlabel('Важность', fontsize=12)
    
    plt.subplot(122)
    plt.barh(y_pos,top_scores, align='center', alpha=0.5)
    plt.title('Релевантные', fontsize=12)
    plt.yticks(y_pos, top_words, fontsize=14)
    plt.suptitle(name, fontsize=16)
    plt.xlabel('Важность', fontsize=12)
    
    plt.subplots_adjust(wspace=0.8)
    plt.show()
````

````python
def train_model(classifier, feature_vector_train, label_train, feature_vector_valid, label_valid):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label_train)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)
    
    print(str(classifier).split('(')[0])
#     print(metrics.accuracy_score(label_valid, predictions))

    report = classification_report(label_valid, predictions)
    print(report)

    cm = metrics.confusion_matrix(y_cats_test, predictions)
    print('Прогноз (сверху-вниз), Истина (слева-направо)')
    print(cm)
    
    this_scores = cross_val_score(classifier, feature_vector_train, label_train, scoring='f1_macro', cv=9, n_jobs=-1)
    scores = np.mean(this_scores)
    scores_std = np.std(this_scores)
    print('Train CV-score %.3f +- %.3f' % (scores, np.array(scores_std)))
    
    this_scores = cross_val_score(classifier, feature_vector_valid, label_valid, scoring='f1_macro', cv=9, n_jobs=-1)
    scores = np.mean(this_scores)
    scores_std = np.std(this_scores)
    print('Test  CV-score %.3f +- %.3f' % (scores, np.array(scores_std)))
    
    print('')
    return classifier
````

## Чистка и стандартизация текста

**Чеклист по очистке данных**

- Удалить все нерелевантные символы (например, любые символы, не относящиеся к цифро-буквенным).
- Токенизировать текст, разделив его на индивидуальные слова.
- Удалить нерелевантные слова — например, упоминания в Twitter или URL-ы.
- Перевести все символы в нижний регистр для того, чтобы слова «привет», «Привет» и «ПРИВЕТ» считались одним и тем же словом.
- Рассмотрите возможность совмещения слов, написанных с ошибками, или имеющих альтернативное написание (например, «круто»/«круть»/ «круууто»)
- Рассмотрите возможность проведения лемматизации, т. е. сведения различных форм одного слова к словарной форме (например, «машина» вместо «машиной», «на машине», «машинах» и пр.)

````python
def standardize_text(df, columns):
    for c in columns:
        df[c] = df[c].str.lower()
        df[c] = df[c].str.replace("ё", "е")
#         df[c] = df[c].str.replace(r"[^А-Яа-я0-9()!?@\'\`\"\_\n]", " ")
        df[c] = df[c].str.replace(r"[^А-Яа-я()!?@\'\`\"\n]", " ")
        
        # Замена нескольких пробелов одним
        df[c] = df[c].str.replace(r"\s+", " ")
    return df



stemmer = SnowballStemmer('russian')

def stem_text(text):
    stem = [stemmer.stem(w) for w in text.split()]
    return ' '.join(stem)

def stem_data(df):
    df['text_stem'] = df.text.apply(stem_text)
    return df



def tokenize_data(df):
    tokenizer = RegexpTokenizer(r'\w+')
#     df["text_tokens"] = df["text"].apply(tokenizer.tokenize)
    df["text_tokens"] = df["text_stem"].apply(tokenizer.tokenize)
    return df



# удаляем шум из данных
def remove_noise(data):
    # пропуски данных
    data.fillna('не определено', inplace=True)

    data.drop(data[data.theme == 'тестовая тема' ].index, inplace=True)
    data.drop(data[data.category == 'тестовая категория' ].index, inplace=True)
    data.drop(data[data.executor == 'тестовый орган власти' ].index, inplace=True)
    
    # удаляем дубликаты
    data.drop_duplicates(subset='text', inplace=True)
    
    # удаляем пустые
    selected = data[data.text.str.len() < 1 ].index
    data.drop(selected, inplace=True)
    
    # короткие-длинные тексты
#     selected = data[data.text.str.len() < 30 ].index   # короткие сообщения. На тесте становится хуже на 5-7%
    selected = data[data.text.str.len() > 900 ].index   # длинные сообщения. Хороший показатель
#    > 500-900    качество лучше чем 900 не становится
    data.drop(selected, inplace=True)
  
    return data



def drop_smalldata(data):
    # где достаточно "много" объектов для КАТЕГОРИИ
    gb = data.groupby('category').count()
    elements = gb[gb[gb.columns[0]] > 50].index.values
    data = data.loc[ data.category.isin(elements)]
    # data.groupby('category').groups  # для проверки, что выбраны нужные группы

    gb = data.groupby('theme').count()
    elements = gb[gb[gb.columns[0]] > 90].index.values   # качество категоризации взлетело до 85-90%
    data = data.loc[ data.theme.isin(elements)]
#     # data.groupby('theme').groups  # для проверки, что выбраны нужные группы

    gb = data.groupby('executor').count()
    elements = gb[gb[gb.columns[0]] > 5].index.values   # качество категоризации 
#     elements = gb[gb[gb.columns[0]] > 29].index.values   # качество категоризации 
#     elements = gb[gb[gb.columns[0]] > 19].index.values   # качество категоризации 
    data = data.loc[ data.executor.isin(elements)]
#     # data.groupby('theme').groups  # для проверки, что выбраны нужные группы

    
    return data



stop_words = set(stopwords.words("russian"))

def remove_stopwords_from_text(text):
    without_sw = [word for word in text if not word in stop_words]
    return without_sw

def remove_stopwords(data):
    data['tokens_wout_stopwords'] = data.text_tokens.apply(remove_stopwords_from_text)
    return data




#  -------------------------------------------------------------------------------------------------------

data = init_data()

# категориальные признаки
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
len(categorical_columns)

data = standardize_text(data, categorical_columns)
data = remove_noise(data)
data = drop_smalldata(data)

data = stem_data(data)
data = tokenize_data(data)
data = remove_stopwords(data)


data.reset_index(inplace=True)
data.drop('index', axis=1, inplace=True)



# data.to_csv("data_cleaned.csv")
print(data.shape)
data.head()
````

````python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - An object to be used as a cross-validation generator.
          - An iterable yielding train/test splits.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Количество примеров")
    plt.ylabel("AUC")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='roc_auc')
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.35,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.35, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Обучающая выборка")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Кроссвалидация")

    plt.legend(loc="best")
    return plt
````
